```python
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
```

### Short Time Fourier Transform - STFT
STFT explains the time frequency of the signal in a more compact form, unlike Discreet Fourier Tranform (DFT), STFT only keeps the real values and discards the imaginary part. We usually take the magnitudes (absolute values) when evaluating the spectrogram. 

```python
data, samplerate = librosa.load('StarWars3.wav', sr=None)   # time domain of signal
plt.plot(data)  
print(data.shape) 

ipd.Audio('StarWars3.wav')      
```

```python
spectro = librosa.core.stft(data, hop_length=513, window='hann', n_fft=1024) # frequency domain of signal
print(spectro.shape)
plt.matshow(abs(spectro), aspect='equal')   # stft magnitudes
plt.show()
```

### STFT - De-beep

```python
# beep sound is shown as a line in the spectrogram

data_x, sr = librosa.load('x-Copy1.wav')     
spectro_x = librosa.core.stft(data_x, hop_length=800, window='hann', n_fft=1600) 
plt.matshow(abs(spectro_x), aspect='auto')   
plt.show()   
```

```python
# remove the beep sound by setting mangitude to zero on the beep frequency

de_beep = spectro_x 
de_beep[140:160] = 0  
plt.matshow(abs(de_beep), aspect='auto')  
```
